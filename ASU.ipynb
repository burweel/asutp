```python
def shiftFunc(df, cols, cols2name, lag=1):
    df = df.copy()
    for col_mean, col2name in zip(cols, cols2name):
        df[col2name] = df[col_mean].shift(lag, fill_value=30)
    return df.copy()
```

```python
# add shift 1
df = shiftFunc(df, 
              ('ai1 mean', 'ai2 mean', 'ai3 mean'),
              ('A1shifted_p1', 'A2shifted_p1', 'A3shifted_p1'),
              lag=1)
```

```python
# add shift 5
df = shiftFunc(df, 
              ('ai1 mean', 'ai2 mean', 'ai3 mean'),
              ('A1shifted_p5', 'A2shifted_p5', 'A3shifted_p5'),
              lag=5)
```

```python
# add shift 20
df = shiftFunc(df, 
              ('ai1 mean', 'ai2 mean', 'ai3 mean'),
              ('A1shifted_p20', 'A2shifted_p20', 'A3shifted_p20'),
              lag=20)
```

```python
df['lop_Ai1_d1_p1'] = df['A1shifted_p1'] - df['ai1 mean']
df['lop_Ai2_d1_p1'] = df['A2shifted_p1'] - df['ai2 mean']
df['lop_Ai3_d1_p1'] = df['A3shifted_p1'] - df['ai3 mean']

df['lop_Ai1_d1_p5'] = (df['A1shifted_p5'] - df['ai1 mean']) / 5
df['lop_Ai2_d1_p5'] = (df['A2shifted_p5'] - df['ai2 mean']) / 5
df['lop_Ai3_d1_p5'] = (df['A3shifted_p5'] - df['ai3 mean']) / 5

df['lop_Ai1_d1_p20'] = (df['A1shifted_p20'] - df['ai1 mean']) / 20
df['lop_Ai2_d1_p20'] = (df['A2shifted_p20'] - df['ai2 mean']) / 20
df['lop_Ai3_d1_p20'] = (df['A3shifted_p20'] - df['ai3 mean']) / 20
```
